```python
#| export
from fasthtml.common import Div, Span

from cjm_fasthtml_tailwind.utilities.spacing import p
from cjm_fasthtml_tailwind.utilities.typography import font_size
from cjm_fasthtml_tailwind.utilities.effects import opacity
from cjm_fasthtml_tailwind.core.base import combine_classes

from cjm_fasthtml_virtual_collection.core.models import VirtualCollectionState
from cjm_fasthtml_virtual_collection.core.html_ids import VirtualCollectionHtmlIds
from cjm_fasthtml_virtual_collection.core.windowing import compute_window
```

```python
#| export
def render_footer(state: VirtualCollectionState,     # Collection state
                  ids: VirtualCollectionHtmlIds,      # HTML IDs
                  oob: bool = False,                  # Whether to include hx-swap-oob
                 ) -> Div:  # Footer element
    """Render the footer with item range indicator."""
    render_start, render_end = compute_window(
        state.window_start, state.visible_rows, state.total_items
    )

    if state.total_items == 0:
        text = "No items"
    else:
        text = f"Showing {render_start + 1}-{render_end} of {state.total_items:,} items"

    footer = Div(
        Span(text),
        id=ids.footer,
        cls=combine_classes(font_size.sm, opacity(70), p.x(2), p.y(1)),
    )

    if oob:
        footer.attrs["hx-swap-oob"] = "outerHTML"

    return footer
```

## Tests

```python
from fasthtml.common import to_xml
from cjm_fasthtml_virtual_collection.core.models import VirtualCollectionState
from cjm_fasthtml_virtual_collection.core.html_ids import VirtualCollectionHtmlIds

ids = VirtualCollectionHtmlIds(prefix="t")
```

```python
# Test normal footer
state = VirtualCollectionState(total_items=500, visible_rows=15, window_start=0)
footer = render_footer(state, ids)
html = to_xml(footer)
assert 'id="t-footer"' in html
assert 'Showing 1-15 of 500 items' in html
print("Normal footer test passed")
```

```python
# Test footer at end
state = VirtualCollectionState(total_items=500, visible_rows=15, window_start=485)
footer = render_footer(state, ids)
html = to_xml(footer)
assert 'Showing 486-500 of 500 items' in html
print("End footer test passed")
```

```python
# Test empty footer
state = VirtualCollectionState(total_items=0, visible_rows=15)
footer = render_footer(state, ids)
html = to_xml(footer)
assert 'No items' in html
print("Empty footer test passed")
```

```python
# Test OOB footer
state = VirtualCollectionState(total_items=500, visible_rows=15, window_start=0)
footer = render_footer(state, ids, oob=True)
html = to_xml(footer)
assert 'hx-swap-oob="outerHTML"' in html
print("OOB footer test passed")
```
